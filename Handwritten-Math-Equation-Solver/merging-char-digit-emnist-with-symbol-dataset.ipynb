```python
import os
import numpy as np
from PIL import Image, ImageFilter

def imageprepare(file_path):
    """
    This function returns the pixel values.
    The input is a png file location.
    """
    im = Image.open(file_path).convert('L')
    width = float(im.size[0])
    height = float(im.size[1])
    newImage = Image.new('L', (28, 28), (255))  # creates white canvas of 28x28 pixels

    if width > height:
        # Width is bigger. Width becomes 20 pixels.
        nheight = int(round((20.0 / width * height), 0))
        if nheight == 0:
            nheight = 1
        img = im.resize((20, nheight), Image.ANTIALIAS).filter(ImageFilter.SHARPEN)
        wtop = int(round(((28 - nheight) / 2), 0))
        newImage.paste(img, (4, wtop))
    else:
        # Height is bigger. Height becomes 20 pixels.
        nwidth = int(round((20.0 / height * width), 0))
        if nwidth == 0:
            nwidth = 1
        img = im.resize((nwidth, 20), Image.ANTIALIAS).filter(ImageFilter.SHARPEN)
        wleft = int(round(((28 - nwidth) / 2), 0))
        newImage.paste(img, (wleft, 4))

    tv = list(newImage.getdata())
    # Normalize pixels to 0 and 1. 0 is pure white, 1 is pure black.
    tva = [(255 - x) * 1.0 / 255.0 for x in tv]
    return tva

# Example usage
math_symbols_dir = "/kaggle/input/handwritten-math-symbols-dataset/images"
X_math_symbols = []
y_math_symbols = []

class_names = os.listdir(math_symbols_dir)
for class_name in class_names:
    class_dir = os.path.join(math_symbols_dir, class_name)
    if os.path.isdir(class_dir):
        for filename in os.listdir(class_dir):
            file_path = os.path.join(class_dir, filename)
            pixel_values = imageprepare(file_path)
            X_math_symbols.append(pixel_values)
            y_math_symbols.append(class_name)

# Convert to numpy arrays
X_math_symbols = np.array(X_math_symbols)
y_math_symbols = np.array(y_math_symbols)

# Now, X_math_symbols contains the pixel values of handwritten math symbols
# y_math_symbols contains the corresponding class labels

# Merge with EMNIST dataset as described earlier
# For merging, you can concatenate X_math_symbols and y_math_symbols with X_emnist and y_emnist respectively
# Ensure that the labels for math symbols don't overlap with EMNIST labels
```

```python
import os
import numpy as np
import pandas as pd
from PIL import Image

# Assuming your data is in a DataFrame called df
df = pd.read_csv("/kaggle/input/emnist/emnist-byclass-train.csv")

# Create a directory to store the images if it doesn't exist
os.makedirs("/kaggle/working/images", exist_ok=True)

# Iterate over each row in the DataFrame
cnt = 0
for index, row in df.iterrows():
    cnt+=1
    if cnt < 5:
        label = str(row[0])  # Assuming the label is in the first column
        image_data = np.array(row[1:], dtype=np.uint8)  # Convert to uint8

        # Reshape the image data into a 2D array (assuming it's flattened)
        image_data = image_data.reshape((28, 28))

        # Create a subfolder for the label if it doesn't exist
        label_dir = os.path.join("images", label)
        if not os.path.exists(label_dir):
            os.makedirs(label_dir)

        # Save the image to the corresponding subfolder
        img = Image.fromarray(image_data)
        img.save(os.path.join(label_dir, f"{index}.png"))
    else:
        break
```
