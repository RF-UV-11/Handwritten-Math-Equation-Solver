```python
import random
import shutil
```

```python
directory = "/kaggle/input/handwritten-math-symbols-dataset/images"
```

```python
desired_sample_images = 1118
```

```python
for folder in ['(',')','+','-','x','']:
    folder_path = os.path.join(directory + '/' + folder)
    if os.path.exists(folder_path):
        files = os.listdir(folder_path)
        
        sampled_files = random.sample(files,min(desired_sample_images,len(files)))
        sampled_folder_path = os.path.join(f'{folder}_sampled')
        os.makedirs(sampled_folder_path,exist_ok=True)
        
        for file in sampled_files:
            file_path = os.path.join(folder_path,file)
            shutil.copy(file_path,sampled_folder_path)
```

```python
import os
import numpy as np
import pandas as pd
from PIL import Image

# Assuming your data is in a DataFrame called df

# Iterate over each row in the DataFrame
for index, row in chopped_df1.iterrows():
    label = str(row[0])  # Assuming the label is in the first column
    image_data = np.array(row[1:], dtype=np.uint8)  # Convert to uint8

    # Reshape the image data into a 2D array (assuming it's flattened)
    image_data = image_data.reshape((28, 28))

    # Create a subfolder for the label if it doesn't exist
    label_dir = os.path.join("/kaggle/working", label)
    if not os.path.exists(label_dir):
        os.makedirs(label_dir)

    # Save the image to the corresponding subfolder
    img = Image.fromarray(image_data)
    img.save(os.path.join(label_dir, f"{index}.png"))

```

```python
import os
import shutil
import zipfile

# Define the folder path to be zipped
folder_path = '/kaggle/working'

# Define the name for the zip file
zip_file_name = 'kaggle_working.zip'

# Create a ZipFile object
with zipfile.ZipFile(zip_file_name, 'w') as zipf:
    # Iterate over all files and subdirectories in the folder
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            # Get the full path of the file
            file_path = os.path.join(root, file)
            # Add the file to the zip archive
            zipf.write(file_path, os.path.relpath(file_path, folder_path))

# Print the path to the created zip file
print("Zip file created:", zip_file_name)
```
